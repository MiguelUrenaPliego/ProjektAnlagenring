```python
import geopandas as gpd 
import pandas as pd 

import plot_helpers
```

```python
pois = pd.read_csv("Bestandsanalyse.csv")
# Rename columns
pois = pois.rename(columns={
    "Breitengrad": "lat",
    "Längengrad": "lon"
})

# Create GeoDataFrame with CRS EPSG:4326
pois = gpd.GeoDataFrame(
    pois,
    geometry=gpd.points_from_xy(pois.lon, pois.lat),
    crs="EPSG:4326"
)
import folium
from folium.plugins import MeasureControl
# Create map centered on POIs
m = folium.Map(
    location=[pois.lat.mean(), pois.lon.mean()],
    zoom_start=15,
    tiles=None,
    control_scale=True
)

# Google Hybrid layer with 50% opacity
folium.TileLayer(
    tiles="https://mt1.google.com/vt/lyrs=y&x={x}&y={y}&z={z}",
    attr="Google",
    name="Google Hybrid",
    overlay=False,
    control=True,
    opacity=0.5
).add_to(m)

# Add POIs
m = plot_helpers.general_map(
    m=m,
    pois=pois,
    poi_column="Typ",
    poi_icon_column="emoji"
)
m.add_child(MeasureControl())
m.save("Bestandsanalyse.html")
```
